# Pooled OLS (Модель пула, модель сквозной регрессии)

```python
# Импортируем класс PooledOLS из модуля linearmodels.panel для построения модели панельных данных методом pooled OLS
from linearmodels.panel import PooledOLS

# Выбираем объясняющие переменные из DataFrame 'growth' и сохраняем их в переменную 'x' (регрессоры модели)
x = df[['cost_res',	'len_roads',	'invest_FA',	'invest_ICT',	'online_sal', 'employ_ICT', 'per_comp', 'employ_gen', 'FA_full_v', 'el_cons']]

# Выбираем зависимую переменную из DataFrame 'growth' и сохраняем ее в переменную 'y'
y = df['GRP_pc']

# Добавляем константу к объясняющим переменным, что необходимо делать построении линейных моделей
x = sm.add_constant(x)
```

```python
# Создаем объект модели PooledOLS, передавая зависимую переменную 'y' и объясняющие переменные 'x'
model_OLS = PooledOLS(y, x)

# Обучаем модель и оцениваем параметры, сохраняя результаты в переменную 'result_OLS'
result_OLS = model_OLS.fit()

# Выводим результаты оценки модели на экран
print(result_OLS)
```

```python
# Создаем объект модели PooledOLS, передавая зависимую переменную 'y' и объясняющие переменные 'x'
model_OLS2 = PooledOLS(y, x)

# Обучаем модель и оцениваем параметры, сохраняя результаты в переменную 'result_OLS'
result_OLS2 = model_OLS.fit(cov_type="robust")

# Выводим результаты оценки модели на экран
print(result_OLS2)
```

# LSDV (least squares dummy variables model, модель с фиктивными переменными)

```python
# Преобразуем категориальные переменные в фиктивные (dummy) переменные
binary_variables = pd.get_dummies(df['region'], prefix='i_')
# Объединяем преобразованные переменные с исходными данными
df = pd.concat([df, binary_variables], axis=1)
df
```

```python
# Выбираем нужные признаки для анализа и добавляем константу к матрице признаков
x_i = sm.add_constant(df[['cost_res',	'len_roads',	'invest_FA',	'invest_ICT',	'online_sal', 'employ_ICT', 'per_comp', 'employ_gen', 'FA_full_v', 'el_cons']])

var_excl = 'i__1'  # Имя переменной, исключаемой из добавления в матрицу признаков

# Проходим по всем фиктивным переменным и добавляем их в матрицу признаков, за исключением переменной var_excl
for var in binary_variables:
    if var != var_excl:
        x_i[var] = binary_variables[var]
```

```python
x_i
```

```python
# Создаем модель пулованных (pooled) МНК с зависимой переменной y и матрицей признаков x_i
model_OLS_i = PooledOLS(y, x_i, check_rank=False)

# Обучаем модель на данных
result_OLS_i = model_OLS_i.fit()

# Выводим результаты оценки модели
print(result_OLS_i)
```

# Fixed effects model (модель с фиксированными эффектами)

```python
from linearmodels.panel import PanelOLS
# Создаем модель фиксированных эффектов (Fixed Effects) с зависимой переменной y и матрицей признаков x,
# включая фиксированные эффекты для панелей
model_FE_i = PanelOLS(y, x, entity_effects=True)

# Обучаем модель на данных
result_FE_i = model_FE_i.fit()

# Выводим результаты оценки модели
print(result_FE_i)
```

```python
# Создаем модель фиксированных эффектов (Fixed Effects) с зависимой переменной y и матрицей признаков x,
# включая фиксированные панельные и временные эффекты
model_FE_it = PanelOLS(y, x, entity_effects=True, time_effects=True)

# Обучаем модель на данных
result_FE_it = model_FE_it.fit()

# Выводим результаты оценки модели
print(result_FE_it)
```

```python
# Создаем модель фиксированных эффектов (Fixed Effects) с зависимой переменной y и матрицей признаков x,
# включая фиксированные панельные и временные эффекты
model_FE_it2 = PanelOLS(y, x, entity_effects=True, time_effects=True)

# Обучаем модель на данных
result_FE_it2 = model_FE_it.fit(cov_type='clustered', cluster_entity=True)

# Выводим результаты оценки модели
print(result_FE_it2)
```

```python
# Создаем модель фиксированных эффектов (Fixed Effects) с зависимой переменной y и матрицей признаков x,
# включая фиксированные панельные и временные эффекты
model_FE_it3 = PanelOLS(y, x, entity_effects=True, time_effects=True)

# Обучаем модель на данных
result_FE_it3 = model_FE_it.fit(cov_type='clustered', cluster_time=True)

# Выводим результаты оценки модели
print(result_FE_it3)
```

# Random effects model (модель со случайными эффектами)

```python
from linearmodels.panel import RandomEffects

# Создаем модель случайных эффектов (Random Effects) с зависимой переменной y и матрицей признаков x
model_RE = RandomEffects(y, x)

# Обучаем модель на данных
result_RE = model_RE.fit()

# Выводим результаты оценки модели
print(result_RE)
```

# Сравнение моделей

```python
from linearmodels.panel import compare

# Сравниваем результаты различных моделей
results_compare = compare({
    'Results of pooled regression': result_OLS,
    'Results of pooled regression2 (robust standart errors)': result_OLS2,
    'Results of LSDV model': result_OLS_i,
    'Results FE model (panel)': result_FE_i,
    'Results FE model (panel+year)': result_FE_it,
    'Results FE model (panel+year clustered entity)': result_FE_it2,
    'Results FE model (panel+year clustered entity+time)': result_FE_it3,
    'Results RE model': result_RE
}, precision='std_errors', stars=True)

# Выводим результаты сравнения моделей
print(results_compare)
```

```python
# Тест Хаусмана
hausman_result = compare({'FE':  result_FE_i, 'RE':  result_RE}, precision='std_errors')
print("Результаты теста Хаусмана:")
print(hausman_result)
```

```python
import numpy as np
from scipy.stats import chi2

# Предположим, что result_FE и result_RE — результаты оценок FE и RE
beta_FE = result_FE_i.params
beta_RE = result_RE.params

# Ковариационные матрицы коэффициентов
cov_FE = result_FE_i.cov
cov_RE = result_RE.cov

# Разница коэффициентов
diff = beta_FE - beta_RE

# Разница ковариационных матриц (и её обращение)
cov_diff = cov_FE - cov_RE
cov_diff_inv = np.linalg.inv(cov_diff)

# Статистика Хаусмана
hausman_stat = diff.T @ cov_diff_inv @ diff
dof = len(beta_FE)  # Число степеней свободы
p_value = 1 - chi2.cdf(abs(hausman_stat), dof)

print(f"Статистика Хаусмана: {abs(hausman_stat)}")
print(f"p-value: {p_value}")
```
